```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.preprocessing import StandardScaler, LabelEncoder, MinMaxScaler
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.svm import SVR
from sklearn.neighbors import KNeighborsRegressor
from sklearn.neural_network import MLPRegressor

```

# 3. Data Pre-processing and Validation

"""

The data pre-processing step includes: conversion of categorical columns into numerical values which is in our case is "purchase_price".

After that we have applied train_test_split to split the dataset into training and testing data. We have splitted the dataset 
into 75% for training and 25% for testing. 

"""

# Dropping null values 

```python
df = df.dropna(axis=0)

df
```

# Converting categorical columns into numerical by using label encoding technique

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder

# Assuming df is your dataframe and these are the remaining categorical columns
categorical_columns = ['house_type', 'sales_type', "city", "area", 'region']

encoded_df = pd.DataFrame()
encoded_df = df

# Apply Label Encoding for ordinal columns
label_encoder = LabelEncoder()

encoded_df[categorical_columns[0]] = label_encoder.fit_transform(encoded_df[categorical_columns[0]])
encoded_df[categorical_columns[1]] = label_encoder.fit_transform(encoded_df[categorical_columns[1]])
encoded_df[categorical_columns[2]] = label_encoder.fit_transform(encoded_df[categorical_columns[2]])
encoded_df[categorical_columns[3]] = label_encoder.fit_transform(encoded_df[categorical_columns[3]])
encoded_df[categorical_columns[4]] = label_encoder.fit_transform(encoded_df[categorical_columns[4]])

encoded_df
```

# Dropping Un-necessary columns from the dataset

```python
df_dropped = encoded_df.drop(['date', 'quarter', 'address', 'zip_code', 'house_id'], axis=1)

df_dropped
```

# Train-Test Split

```python
X = df_dropped.drop('purchase_price', axis=1)
y = df_dropped['purchase_price']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

# Scaling the dataset to ensure the best accuracy of the model

```python
scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

# Calculate the correlation matrix

```python
corr_matrix = df_dropped.corr()

# Print the correlation matrix
corr_matrix
```

```python
# Saving cleaned dataset
df_dropped.to_csv('DK_Housing_Prices_Cleaned.csv')
```

# !-------------- 4. Model Training and Evaluation ------------------!
We have taken 6 models: Linear Regression, Ridge, Lasso, Random Forest regressor, Gradient Boosted regressor and K nearest neighbour for the regression purpose and to get the best accuracy out of these 6 models. 

# Training our models

```python
# Linear Regression
linear_reg = LinearRegression()
linear_reg.fit(X_train_scaled, y_train)
linear_reg_pred = linear_reg.predict(X_test_scaled)
print("\nLinear Regressor Metrics:")
print("Mean Absolute Error:", mean_absolute_error(y_test, linear_reg_pred))
print("Mean Squared Error:", mean_squared_error(y_test, linear_reg_pred))
print("Root Mean Squared Error:", np.sqrt(mean_squared_error(y_test, linear_reg_pred)))


# Ridge Regression
ridge_reg = Ridge(alpha=1.0)
ridge_reg.fit(X_train_scaled, y_train)
ridge_reg_pred = ridge_reg.predict(X_test_scaled)
print("\nRidge Regressor Metrics:")
print("Mean Absolute Error:", mean_absolute_error(y_test, ridge_reg_pred))
print("Mean Squared Error:", mean_squared_error(y_test, ridge_reg_pred))
print("Root Mean Squared Error:", np.sqrt(mean_squared_error(y_test, ridge_reg_pred)))


# Lasso Regression
lasso_reg = Lasso(alpha=1.0)
lasso_reg.fit(X_train_scaled, y_train)
lasso_reg_pred = lasso_reg.predict(X_test_scaled)
print("\nLasso Regressor Metrics:")
print("Mean Absolute Error:", mean_absolute_error(y_test, lasso_reg_pred))
print("Mean Squared Error:", mean_squared_error(y_test, lasso_reg_pred))
print("Root Mean Squared Error:", np.sqrt(mean_squared_error(y_test, lasso_reg_pred)))


# Random Forest Regression
rf_reg = RandomForestRegressor()
rf_reg.fit(X_train_scaled, y_train)
rf_reg_pred = rf_reg.predict(X_test_scaled)
print("\nRandom Forest Regressor Metrics:")
print("Mean Absolute Error:", mean_absolute_error(y_test, rf_reg_pred))
print("Mean Squared Error:", mean_squared_error(y_test, rf_reg_pred))
print("Root Mean Squared Error:", np.sqrt(mean_squared_error(y_test, rf_reg_pred)))
```

```python
# Gradient Boosting Regressor
gbr = GradientBoostingRegressor()
gbr.fit(X_train, y_train)
y_pred_gbr = gbr.predict(X_test)
print("Gradient Boosting Regressor Metrics:")
print("Mean Absolute Error:", mean_absolute_error(y_test, y_pred_gbr))
print("Mean Squared Error:", mean_squared_error(y_test, y_pred_gbr))
print("Root Mean Squared Error:", np.sqrt(mean_squared_error(y_test, y_pred_gbr)))
```

```python
# K-Nearest Neighbors (KNN) Regressor
knn = KNeighborsRegressor(n_neighbors=5)
knn.fit(X_train, y_train)
y_pred_knn = knn.predict(X_test)
print("\nK-Nearest Neighbors Regressor Metrics:")
print("Mean Absolute Error:", mean_absolute_error(y_test, y_pred_knn))
print("Mean Squared Error:", mean_squared_error(y_test, y_pred_knn))
print("Root Mean Squared Error:", np.sqrt(mean_squared_error(y_test, y_pred_knn)))
```

# Calculating R2 score


```python
# Linear Regression R2 Score
linear_reg_r2 = r2_score(y_test, linear_reg.predict(X_test_scaled))
print(f"Linear Regression R2 Score: {linear_reg_r2:.4f}")

# Ridge Regression R2 Score
ridge_reg_r2 = r2_score(y_test, ridge_reg.predict(X_test_scaled))
print(f"Ridge Regression R2 Score: {ridge_reg_r2:.4f}")

# Lasso Regression R2 Score
lasso_reg_r2 = r2_score(y_test, lasso_reg.predict(X_test_scaled))
print(f"Lasso Regression R2 Score: {lasso_reg_r2:.4f}")

# RF Regression R2 Score
rf_reg_r2 = r2_score(y_test, rf_reg.predict(X_test_scaled))
print(f"Random Forest Regression R2 Score: {rf_reg_r2:.4f}")



# Gradient Boosted Regression R2 Score
gbr_reg_r2 = r2_score(y_test, y_pred_gbr) 
print(f"Gradient Boosted Regression R2 Score: {gbr_reg_r2:.4f}")

# K nearest neighbour Regression R2 Score
knn_reg_r2 = r2_score(y_test, y_pred_knn)
print(f"K nearest neighbour Regression R2 Score: {knn_reg_r2:.4f}")

```
